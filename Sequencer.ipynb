# Sequencer

```python
def sequencing(sim_data, new_seq):
    '''This function converts BLL parsed data into sequences built in the following way:
        {'state': {'wavelength', 'solution', 'width', 'concentration', 'absorbance'}, 
           'actions': {'action', 'label', 'timestamp'}}'''
    
    for i in range(len(sim_data['logs'])):
        action = sim_data['logs'][i]['action']
        timestamp = sim_data['logs'][i]['timestamp']

        if action == 'open_instruction' or action == 'close_instruction':
            new_seq['actions']['action'].append(action)
            new_seq['actions']['label'].append('instruction')
            new_seq['actions']['timestamp'].append(timestamp)

            new_seq['state']['width'].append(new_seq['state']['width'][-1])
            new_seq['state']['wavelength'].append(new_seq['state']['wavelength'][-1])
            new_seq['state']['solution'].append(new_seq['state']['solution'][-1])
            new_seq['state']['concentration'].append(new_seq['state']['concentration'][-1])
            new_seq['state']['absorbance'].append(new_seq['state']['absorbance'][-1])
        
        elif action == 'width' or action == 'concentration' or action == 'wavelength':
            new_seq['actions']['action'].append(action)
            new_seq['actions']['label'].append('explore')
            new_seq['actions']['timestamp'].append(timestamp)
            
            if action == 'width':
                new_seq['state']['wavelength'].append(new_seq['state']['wavelength'][-1])
                new_seq['state']['solution'].append(new_seq['state']['solution'][-1])
                new_seq['state']['concentration'].append(new_seq['state']['concentration'][-1])

                if new_seq['state']['width'][-1] == sim_data['logs'][i]['old_value']:
                    new_seq['state']['width'].append(sim_data['logs'][i]['new_value'])
                else:
                    print(i)
                    print("width: previous value is not old_value")
                    break
                
                if 'old_abs' in sim_data['logs'][i]:
                    if new_seq['state']['absorbance'][-1] == sim_data['logs'][i]['old_abs']:
                        new_seq['state']['absorbance'].append(sim_data['logs'][i]['new_abs'])
                    else:
                        print(i)
                        print("width: previous absorbance is not old_abs")
                        break
                else:
                    new_seq['state']['absorbance'].append(new_seq['state']['absorbance'][-1]) 
                
            elif action == 'concentration':
                new_seq['state']['wavelength'].append(new_seq['state']['wavelength'][-1])
                new_seq['state']['solution'].append(new_seq['state']['solution'][-1])
                new_seq['state']['width'].append(new_seq['state']['width'][-1])

                if new_seq['state']['concentration'][-1] == sim_data['logs'][i]['old_value']:
                    new_seq['state']['concentration'].append(sim_data['logs'][i]['new_value'])
                else:
                    print(i)
                    print("concentration: previous value is not old_value")
                    break
                    
                if 'old_abs' in sim_data['logs'][i]:
                    if new_seq['state']['absorbance'][-1] == sim_data['logs'][i]['old_abs']:
                        new_seq['state']['absorbance'].append(sim_data['logs'][i]['new_abs'])
                    else:
                        print(i)
                        print("concentration: previous absorbance is not old_abs")
                        break
                else:
                    new_seq['state']['absorbance'].append(new_seq['state']['absorbance'][-1])
                    
            elif action == 'wavelength':
                new_seq['state']['concentration'].append(new_seq['state']['concentration'][-1])
                new_seq['state']['solution'].append(new_seq['state']['solution'][-1])
                new_seq['state']['width'].append(new_seq['state']['width'][-1])

                if new_seq['state']['wavelength'][-1] == sim_data['logs'][i]['old_value']:
                    new_seq['state']['wavelength'].append(sim_data['logs'][i]['new_value'])
                else:
                    print(i)
                    print("wavelength: previous value is not old_value")
                    break
                    
                if 'old_abs' in sim_data['logs'][i]:
                    if new_seq['state']['absorbance'][-1] == sim_data['logs'][i]['old_abs']:
                        new_seq['state']['absorbance'].append(sim_data['logs'][i]['new_abs'])
                    else:
                        print(i)
                        print("wavelength: previous absorbance is not old_abs")
                        break
                else:
                    new_seq['state']['absorbance'].append(new_seq['state']['absorbance'][-1])
            
        elif action == 'solution':
            new_seq['actions']['action'].append(action)
            new_seq['actions']['label'].append('explore')
            new_seq['actions']['timestamp'].append(timestamp)
            
            new_seq['state']['width'].append(new_seq['state']['width'][-1])

            if new_seq['state']['solution'][-1] == sim_data['logs'][i]['old_value']:
                new_seq['state']['solution'].append(sim_data['logs'][i]['new_value'])
            else:
                print(i)
                print(new_seq['state']['solution'][-1])
                print("solution: previous value is not old_value")
                break
                
            if 'old_conc' in sim_data['logs'][i]:
                if new_seq['state']['concentration'][-1] == sim_data['logs'][i]['old_conc']:
                    new_seq['state']['concentration'].append(sim_data['logs'][i]['new_conc'])
                else:
                    print(i)
                    print("solution: previous concentration is not old_conc")
                    break
            else:
                new_seq['state']['concentration'].append(new_seq['state']['concentration'][-1])
                
            if 'old_wave' in sim_data['logs'][i]:    
                if new_seq['state']['wavelength'][-1] == sim_data['logs'][i]['old_wave']:
                    new_seq['state']['wavelength'].append(sim_data['logs'][i]['new_wave'])
                else:
                    print(i)
                    print("solution: previous wavelength is not old_wave")
                    break
            else:
                new_seq['state']['wavelength'].append(new_seq['state']['wavelength'][-1])
                
            if 'old_abs' in sim_data['logs'][i]:    
                if new_seq['state']['absorbance'][-1] == sim_data['logs'][i]['old_abs']:
                    new_seq['state']['absorbance'].append(sim_data['logs'][i]['new_abs'])
                else:
                    print(i)
                    print("solution: previous absorbance is not old_abs")
                    break
            else:
                new_seq['state']['absorbance'].append(new_seq['state']['absorbance'][-1])
            
        elif action == 'open_solution_menu' or action == 'ruler' or action == 'record' or action == 'xaxis' or action == 'yaxis' or action == 'xaxis_scale' or action == 'yaxis_scale' or action == 'delete_point_from_table' or action == 'move_down_point' or action == 'move_up_point' or action == 'plot' or action == 'remove_from_graph':
            new_seq['state']['width'].append(new_seq['state']['width'][-1])
            new_seq['state']['wavelength'].append(new_seq['state']['wavelength'][-1])
            new_seq['state']['solution'].append(new_seq['state']['solution'][-1])
            new_seq['state']['concentration'].append(new_seq['state']['concentration'][-1])
            new_seq['state']['absorbance'].append(new_seq['state']['absorbance'][-1])
            
            if action == 'open_solution_menu' or action == 'ruler':
                new_seq['actions']['timestamp'].append(timestamp)
                new_seq['actions']['action'].append(action)
                new_seq['actions']['label'].append('other')

            elif action == 'record' or action == 'xaxis' or action == 'yaxis' or action == 'xaxis_scale' or action == 'yaxis_scale':
                new_seq['actions']['timestamp'].append(timestamp)
                new_seq['actions']['action'].append(action)
                new_seq['actions']['label'].append(action)
                
            elif action == 'delete_point_from_table' or action == 'move_down_point' or action == 'move_up_point':
                new_seq['actions']['timestamp'].append(timestamp)
                new_seq['actions']['action'].append(action)
                new_seq['actions']['label'].append('table')
                
            elif action == 'plot' or action == 'remove_from_graph':
                new_seq['actions']['timestamp'].append(timestamp)
                new_seq['actions']['action'].append(action)
                new_seq['actions']['label'].append('graph')
                
        elif action == 'restore_sim':
            new_seq['actions']['timestamp'].append(timestamp)
            new_seq['actions']['action'].append(action)
            new_seq['actions']['label'].append(action)
            
            count_record = 0

            for j in range(len(new_seq['actions']['action'])):
                if new_seq['actions']['action'][j] == 'record':
                    count_record += 1
                    if count_record == sim_data['logs'][i]['table_point']:
                        new_seq['state']['width'].append(new_seq['state']['width'][j])
                        new_seq['state']['wavelength'].append(new_seq['state']['wavelength'][j])
                        new_seq['state']['solution'].append(new_seq['state']['solution'][j])
                        new_seq['state']['concentration'].append(new_seq['state']['concentration'][j])
                        new_seq['state']['absorbance'].append(new_seq['state']['absorbance'][j])
                        break
                        
        elif action == 'end':
            new_seq['actions']['timestamp'].append(timestamp)
            new_seq['actions']['action'].append(action)
            new_seq['actions']['label'].append(action)
    
    #print('End')
    
    #return new_seq
    
    new_seq_data = {'group': sim_data['group'], 'subgroup': sim_data['subgroup'], 
                    'session_code': sim_data['session_code'], 'task': sim_data['task'],
                   'new_seq': new_seq}
    
    return new_seq_data
```
